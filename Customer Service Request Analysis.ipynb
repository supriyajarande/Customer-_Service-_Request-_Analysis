# 2 Perform basic data exploratory analysis

```python
# 2(i)Utilize missing value treatment

#Drop the row which has all null value
df.dropna(how='all')
```

```python
# Drop the cell which has null value
df.dropna(subset=['Created Date','Closed Date'])
```

```python
#drop the column which has all null value
df1 = df.dropna(how='all', axis=1)
print(df1)
```

```python
# Handle missing value by filling with 0
df2=df1.fillna(value = 0)
df2
```

```python
# to check whether null value still in dataset
df2.isnull().sum()
```

```python

df2["Created Date"] = pd.to_datetime(df2["Created Date"])
df2["Closed Date"] = pd.to_datetime(df2["Closed Date"])
```

```python
df2.dtypes
```

```python
df2["Created Date"].value_counts()[:3]
```

```python
# 2(ii) Analyze the date column and remove the entries if it has an incorrect timeline
df3 = pd.to_datetime(df2['Created Date'],format="%m/%d/%Y %H:%M:%S")
df4 = pd.to_datetime(df2['Closed Date'],format="%m/%d/%Y %H:%M:%S")
df4
```

```python
#2(iii)Draw a frequency plot for city-wise complaints
from matplotlib.pyplot import figure 
figure(figsize=(10,3 ), dpi=80) 

#fd1= df.groupby(["City"]).size().sort_values(ascending=False)[1:]
fd1 = df2["City"].value_counts()[0:10]
print(fd1.head())
fd1.plot();
```

```python
#2(iii)Draw a frequency plot for city-wise complaints
City_Name = df2.groupby(["City"]).size().sort_values(ascending=False)[0:]
print(City_Name.head())
City_Name.plot();
```
